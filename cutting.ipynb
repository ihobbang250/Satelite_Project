```python
import pandas as pd
import os
from shutil import copyfile

# Load the CSV file into a DataFrame
df = pd.read_csv('./data/train.csv')

# Get the first 100 rows from the DataFrame
subset_df = df.head(100)

df.head()
# Extract the IDs and file paths from the subset DataFrame
ids = subset_df['img_id'].tolist()
file_paths = subset_df['img_path'].tolist()

# Create a new directory to store the extracted files
output_dir = './subset_files'
os.makedirs(output_dir, exist_ok=True)

# Copy the corresponding files to the output directory
for id, file_path in zip(ids, file_paths):
    file_name = os.path.basename(file_path)
    output_path = os.path.join(output_dir, file_name)
    copyfile(file_path, output_path)

# Print the extracted IDs and corresponding file paths
for id, file_path in zip(ids, file_paths):
    print(f"ID: {id}, File Path: {file_path}")

subset_df.to_csv('./data/subset.csv', index=False)
```
